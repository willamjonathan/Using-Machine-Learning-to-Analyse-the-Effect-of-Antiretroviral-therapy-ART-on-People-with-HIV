```python
# import modules
import pandas as pd
```

```python
#reading the data from the csv file
df = pd.read_csv('HIV-overyear-cleaned.csv')
df.info()
```

```python
# checking the number of location in each period uptil 2011
d1 = df.loc[(df["Period"] == 2021) , ["Period"]]
d1.info()
d2 = df.loc[(df["Period"] == 2020) , ["Period"]]
d2.info()
d3 = df.loc[(df["Period"] == 2019) , ["Period"]]
d3.info()
d4 = df.loc[(df["Period"] == 2018) , ["Period"]]
d4.info()
d5 = df.loc[(df["Period"] == 2017) , ["Period"]]
d5.info()
d6 = df.loc[(df["Period"] == 2016) , ["Period"]]
d6.info()
d7 = df.loc[(df["Period"] == 2015) , ["Period"]]
d7.info()
d8 = df.loc[(df["Period"] == 2014) , ["Period"]]
d8.info()
d9 = df.loc[(df["Period"] == 2013) , ["Period"]]
d9.info()
d10 = df.loc[(df["Period"] == 2012) , ["Period"]]
d10.info()
d11 = df.loc[(df["Period"] == 2011) , ["Period"]]
d11.info()
```

```python
# cutting the data based on the information above
# tail and head function is used just to check whether i put the correct period or not

df2021 = df.iloc[0:138,:] # 138 data
# df2021.tail()
df2020 = df.iloc[138:277,[1,2,3]] # 139 data
# df2020.tail()
df2019 = df.iloc[277:417,[1,2,3]] # 140 data
# df2019.tail()
df2018 = df.iloc[417:557,[1,2,3]] # 140 data
# df2018.tail()
df2017 = df.iloc[557:697,[1,2,3]] # 140 data
# df2017.tail()
df2016 = df.iloc[697:837,[1,2,3]] # 140 data
# df2016.tail()
df2015 = df.iloc[837:977,[1,2,3]] # 140 data
# df2015.tail()
df2014 = df.iloc[977:1117,[1,2,3]] # 140 data
# df2014.head()
# df2014.tail()
df2013 = df.iloc[1117:1257,[1,2,3]] # 140 data
# df2013.tail()
# df2013.head()
df2012 = df.iloc[1257:1397,[1,2,3]] # 140 data
# df2012.head()
# df2012.tail()
df2011 = df.iloc[1397:1537,[1,2,3]] # 140 data
# df2021.head()
# df2011.tail()
```

```python
# Merging 3 dataframes by using location as the index
# The location number will follow the first df 
dfx= pd.merge(pd.merge(df2021,df2020
                       ,on='Location'),df2019,on='Location')
dfx.rename(columns = {
    
  'FactValueNumeric_x':'FValue2021'
  ,'FactValueNumeric_y':'FValue2020'
  ,'FactValueNumeric' : 'FValue2019'
  ,'Period_x': 'Period2021'
  ,'Period_y': 'Period2020'
  ,'Period': 'Period2019'
  
  }, inplace = True)
# dfx.info()

dfy= pd.merge(pd.merge(dfx,df2018
                       ,on='Location'),df2017,on='Location')
dfy.rename(columns = {
    
  'FactValueNumeric_x':'FValue2018'
  ,'FactValueNumeric_y':'FValue2017'
  ,'Period_x': 'Period2018'
  ,'Period_y': 'Period2017'
  
  
  }, inplace = True)

# dfy.info()
# dfy.head()

dfz= pd.merge(pd.merge(dfy,df2016
                       ,on='Location'),df2015,on='Location')
dfz.rename(columns = {
    
  'FactValueNumeric_x':'FValue2016'
  ,'FactValueNumeric_y':'FValue2015'
  ,'Period_x': 'Period2016'
  ,'Period_y': 'Period2015'
  
  
  }, inplace = True)

# dfz.info()
# dfz.head()

dfyy= pd.merge(pd.merge(dfz,df2014
                       ,on='Location'),df2013,on='Location')
dfyy.rename(columns = {
    
  'FactValueNumeric_x':'FValue2014'
  ,'FactValueNumeric_y':'FValue2013'
  ,'Period_x': 'Period2014'
  ,'Period_y': 'Period2013'
  
  
  }, inplace = True)

# dfyy.info()

dfxx= pd.merge(pd.merge(dfyy,df2012
                       ,on='Location'),df2011,on='Location')
dfxx.rename(columns = {
    
  'FactValueNumeric_x':'FValue2012'
  ,'FactValueNumeric_y':'FValue2011'
  ,'Period_x': 'Period2012'
  ,'Period_y': 'Period2011'
  
  
  }, inplace = True)

dfxx.info()
```
